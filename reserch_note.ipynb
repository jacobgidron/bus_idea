```python
from pyproj import Transformer

# # Define source and target CRS (EPSG codes)
# itm_epsg = 2039  # Israeli ITM (also known as IS:2039)
# wgs84_epsg = 4326 # WGS 84 Geographic
# transformer = Transformer.from_crs(itm_epsg, wgs84_epsg)
# transformer.transform(202679,738786)


crs_from = 'epsg:3857'
crs_to = 'epsg:4326'

# Create a Transformer object for the conversion
# always_xy=True ensures output is always (x, y) or (longitude, latitude)
transformer = Transformer.from_crs(crs_from, crs_to, always_xy=True)

transformer.transform(3886611.373224132,3820674.3162589082)
```

```python
l = []
l2 = []
for i in j["trips"]['33480003_060825']["stop_times"]:
    # "Latitude-longitude information"
    lon,lat = transformer.transform(i["stop_lon"],i["stop_lat"])
    l.append({
        "name": i["stop_name"]["en"],
        "WKT": f"POINT({lon} {lat})"
        # "latitude-longitude": f"{i["stop_lat"]},{i["stop_lon"]}"
    })
    l2.append(f"{lon} {lat}")
    # l2.append({
    #     "name": i["stop_name"]["en"],
    #     "latitude": lat,
    #     "longitude": lon
    #     # "latitude-longitude": f"{i["stop_lat"]},{i["stop_lon"]}"
    # })
```

```python
import csv

with open('stops_64.csv', 'w', newline='') as csvfile:
    # fieldnames = ["name","latitude-longitude"]
    # fieldnames = ["name","latitude","longitude"]
    fieldnames = ["WKT","name"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for x in l:
        writer.writerow(x)
    line = {"name": "line",
        "WKT": "LINESTRING(" + ", ".join(l2) + ")"
        }
    writer.writerow(line)
```
